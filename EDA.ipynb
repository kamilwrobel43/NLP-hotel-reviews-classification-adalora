```python

import torch 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filenames = ['data/tripadvisor_hotel_reviews.csv']
X_col = "Review"
y_col = "Rating"
```

```python
df = pd.read_csv(filenames[0])
df[y_col].value_counts()
df["review_len"] = df.apply(lambda row: len(str(row[X_col]).split()), axis=1)
plt.title("Ratings distribution")
sns.countplot(x=y_col, data=df)
plt.show()
plt.title("Reviews words count distribution")
sns.histplot(x="review_len", data=df)
plt.show()
print(f"n_samples: {len(df)}")
df.describe().T
```
